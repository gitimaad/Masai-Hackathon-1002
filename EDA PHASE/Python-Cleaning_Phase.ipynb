```python
#Importing files required for cleaning of outliers
import pandas as pd
import numpy as np
```

```python
#Reading the dataset we have to remove outliers from and clean the data
train = pd.read_csv('train.csv')
```

```python
#Only outliers replacing is required as the data set has no null values and duplicate values as studied in the EDA.
# Function to detect and replace outliers using the IQR method
def replace_outliers_iqr(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Replace outliers with the median of the column
        median = df[col].median()
        df[col] = np.where((df[col] < lower_bound) | (df[col] > upper_bound), median, df[col])
    return df

# Identifying numeric columns for outlier detection
numeric_columns = train.select_dtypes(include=['int64', 'float64']).columns

# Replacing outliers
data_cleaned = replace_outliers_iqr(train.copy(), numeric_columns)

# Checking the changes
data_cleaned.describe()
```

```python
#Extracting the cleaned dataset which would be used for ML Model.
data_cleaned.to_csv('cleaned_file.csv', index=False)
```
